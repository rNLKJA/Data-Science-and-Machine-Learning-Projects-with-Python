## Omicron Sentiment Analysis using Python

The dataset that I am using for the task of Omicron sentiment analysis is downloaded from Kaggle, which was initially collected from Twitter when people were sharing their opinions about the Omicron variant. So letâ€™s start the task of Omicron sentiment analysis by importing the necessary Python libraries and the dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# data set from kaggle https://www.kaggle.com/gpreda/omicron-rising
data = pd.read_csv("omicron.csv")
data.tail()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
import nltk
import re

nltk.download("stopwords")
stemmer = nltk.SnowballStemmer("english")

from nltk.corpus import stopwords
import string

stopword = set(stopwords.words('english'))
```

```python
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
```

```python
data.text = data.text.apply(clean)
```

```python
text = " ".join(i for i in data.text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = " ".join(i for i in data.hashtags)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
nltk.download("vader_lexicon")
sentiments = SentimentIntensityAnalyzer()
```

```python
scores = data.agg(lambda x: sentiments.polarity_scores(x.text), axis=1)
```

```python
scores = scores.apply(pd.Series)
scores.columns = ['Negative', 'Neutral', 'Positive', 'Compound']
scores.tail()
```

```python
data = pd.concat([data, scores], axis=1)
```

```python
x = sum(data["Positive"])
y = sum(data["Negative"])
z = sum(data["Neutral"])

def sentiment_score(a, b, c):
    if (a>b) and (a>c):
        print("Positive ðŸ˜Š ")
    elif (b>a) and (b>c):
        print("Negative ðŸ˜  ")
    else:
        print("Neutral ðŸ™‚ ")
sentiment_score(x, y, z)
```
